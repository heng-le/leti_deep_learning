```python
import pandas as pd 
import RNA
```

```python
df = pd.read_csv("./data/p04_clinvar_summary.csv")
```

```python
df = df.drop(['Unnamed: 0','Max', 'Indel', '5_homology', 'id', 'name', 'target','snp_id', 'sublibrary','polyT', 'bc1', 'target_seq_adj', 'bc2', '3_homology', 'full_plasmid', 'full_plasmid_length', 'sample', 'total_reads', 'Pos', 'Eff'], axis=1)

df.head()
```

```python
df.shape
```

```python
df_long = pd.melt(df, id_vars=['spacer_add_g', 'spacer_length', 'scaffold', 'target_seq'], 
                  var_name='editing_position', value_name = 'Eff')
df_long.shape
```

```python
df_long.head()
```

```python
df_long['spacer_add_g'] = df_long['spacer_add_g'].apply(lambda x: x[1:])
```

```python
df_long['sec_struc'] = df_long['spacer_add_g'] + df_long['scaffold']
```

```python
df_long['editing_position'] = df_long['editing_position'].astype(str).astype(int)

df_long = df_long[df_long['editing_position'] >= -25]
```

```python
df_long['spacer_add_g'] = df_long['spacer_add_g'].apply(lambda x: x.upper())
df_long['sec_struc'] = df_long['sec_struc'].apply(lambda x: x.upper())
df_long['target_seq'] = df_long['target_seq'].apply(lambda x: x.upper())
```

```python
df_long.head()
```

```python
df_long['sec_struc'].iloc[0]
```

```python
def compute_dot_bracket(row):
    fc = RNA.fold_compound(row['sec_struc'])
    return fc.mfe()[0]

df_long['dot_bracket'] = df_long.apply(compute_dot_bracket, axis=1)
```

```python
df_long.head()
```

```python
df_long.to_csv('./data/p04_clinvar_summary_processed_target.csv', index=False)
```
